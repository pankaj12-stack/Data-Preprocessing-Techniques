```python
"""
    Narrowing down the dataset to make it more easy to plot.

"""

task2_db=df.query('Age > 24 and Age <35')

"""
    Query the columns of a DataFrame with a boolean expression.


    Parameters
    ----------
    expr:str

    Returns
    -------
    DataFrame
        DataFrame resulting from the provided query expression

"""
print(task2_db)
```

```python
"""
    Replacing 0 Values
    There are a lot of 0 values in weight which is not possible as no player cannot have 0 weight so
    we are replacing the 0 values using quantile function

"""
q_low = task2_db["Weight"].quantile(0.01)

"""
    Return values at the given quantile over requested axis.

    Parameters
    ----------
    q:float or array-like, default 0.5 (50% quantile)
        The column name or column position to be used as horizontal coordinates for each point.

    Returns
    -------
    Series or DataFrame
        If q is an array, a DataFrame will be returned where the
        index is q, the columns are the columns of self, and the values are the quantiles.

"""

q_hi  = task2_db["Weight"].quantile(0.95)

task2_db = task2_db[(task2_db["Weight"] < q_hi) & (task2_db["Weight"] > q_low)]
```

```python
"""
    Generating Missing data

    Here we would be generating missing data, 10% to be exact, using sample() function.
    By using this function random rows would be selected and 10% of the total rows would be chaned to NaN

"""
task2_db.loc[task2_db.sample(frac=0.1).index.tolist(), 'Height'] = np.nan

"""
    Return a random sample of items from an axis of object.

    Parameters
    ----------
    frac: float, optional
        Fraction of axis items to return. Cannot be used with n.


    Returns
    -------
    Series or DataFrame
        A new object of same type as caller containing n items randomly sampled from the caller object.


"""

task2_db.loc[task2_db.sample(frac=0.1).index.tolist(), 'Weight'] = np.nan
```

```python
fig, axs, = plt.subplots(2,1,figsize=(16,20))

axs = axs.flatten()
task2_db.index = range(task2_db.shape[0])
print(task2_db.index)


task2_db['Height'].plot(ax=axs[0], title='Height', xlabel='Players')
task2_db['Weight'].plot(ax=axs[1], title='Weight', xlabel='Players')

"""
    As we can see blanks in the line plot we conclude that out data consists of missing values.
    In the next steps we would be dropping the rows which have missing values.

"""
```

```python
val_cols_names = ['Height','Weight']


def plot_series_side_by_side(original, resulting, columns):
    """
    Plots modified and orignial line plot graphs

    Parameters
    ----------
    original: Series or DataFrame
        The object for which the method is called.

    resulting: Series or DataFrame
        The object for which the method is called.

    val_cols_names: column label, or list of labels
         coloumn label or list of labels may be passed


    Returns
    -------

    matplotlib.axes.Axes

    """
    fig, axs = plt.subplots(len(columns),1,figsize=(16,20))
    axs = axs.flatten()
    custom_xlim = (original.index.min(), original.index.max())
    i = 0
    for c in columns:
        resulting[c].plot(ax=axs[i], title=c, xlabel='', xlim=custom_xlim, label='Modified', lw=4, color='red')
        original[c].plot(ax=axs[i], title=c, xlabel='', xlim=custom_xlim, label='Original', color='black')
        axs[i].legend()
        i+=1


drop_df = task2_db.copy()
drop_df = drop_df.dropna()
drop_df = drop_df.dropna(thresh=2)
plot_series_side_by_side(task2_db, drop_df,val_cols_names)

task2_db=drop_df
```

Here we dropped the rows that consisted the missing data and plotted both, the original along with the modified graphs side by side to see the difference.

```python
"""
    Generating Outliers

    Here we would be generating outliers in our dataframe, 5% to be exact, using sample() function.
    By using this function random five percentage of rows would be selected and their vlaues
    would be replaced by random number between 30 and 40

"""

task2_db.loc[task2_db.sample(frac=0.05).index.tolist(), 'Stamina'] = np.random.randint(30,40)
task2_db.loc[task2_db.sample(frac=0.05).index.tolist(), 'Strength'] = np.random.randint(30,40)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,4))
axs = axs.flatten()

task2_db.boxplot(ax=axs[0],column='Stamina', figsize=(10, 8))

"""
    Make a box plot of the DataFrame columns.

    Parameters
    ----------
    by: str or sequence
        Column in the DataFrame to group by.


    Returns
    -------
    matplotlib.axes.Axes or numpy.ndarray of them


"""

task2_db.boxplot(ax=axs[1],column='Strength', figsize=(10, 8))
```

Here we can see the dots on our box plots, there dots are called as outliers. This means that we were successfull in creating outliers in our dataframe.

```python
val= ['Stamina','Strength']

def replace_outliers_IQR(df, columns):
    """
    Replaces outliers using the IQR Method

    Parameters
    ----------
    df: Series or DataFrame
        The object for which the method is called.

    columns: column label, or list of labels
         coloumn label or list of labels may be passed


    Returns
    -------

        df: Series or DataFrame
            Generates dataframe without any outliers

    """
    for c in columns:
        q1=df[c].quantile(0.25)
        q3=df[c].quantile(0.75)
        iqr = q3-q1
        low_lim = q1 - 1.5*iqr
        high_lim = q3 + 1.5*iqr
        print(f"High ({c}):", high_lim)
        print(f"Low ({c}):", low_lim)
        df[c] = np.where(df[c] >= high_lim, high_lim, np.where(df[c] <= low_lim, low_lim, df[c]))

task2_db_outlier = task2_db.copy()
replace_outliers_IQR(task2_db_outlier, val)
plot_series_side_by_side(task2_db, task2_db_outlier, val)
task2_db=task2_db_outlier
```

Here we dropped the rows that consisted the outliers and plotted both, the original along with the modified graphs side by side to see the difference.
